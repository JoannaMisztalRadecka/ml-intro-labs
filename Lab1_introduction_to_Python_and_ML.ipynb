# Working with data and machine learning problems

Nowadays, machine learning algorithms have applications to many real-world problems that would be difficult to solve manually. In the labs, we will learn how to use statistical and machine learning methods to solve problems based on data. 

Before you start building a model, you should define:

* The **task** - what problem are you trying to solve?
* The **dataset** - what data can you use to learn the model how to solve this problem?
* The performance **measure** - how are you going to evaluate your results?

To get an idea of the type of problems, datasets and metrics that can be used, go to the Kaggle competition website https://www.kaggle.com/competitions and explore the descriptions of the tasks from the list.
